# Mais sobre o return

### Pontos importantes:

- Podemos usar no return praticamente qualquer tipo de objeto: (número, string, lista, tupla, dicionário, outros objetos, etc.)
- O return, se for executado, encerra a função, mesmo que dentro dela haja um loop.

```python
#retornar um número
def minha_soma(num1, num2, num3):
    return num1 + num2 + num3

#retornar um texto
def padronizar_texto(texto):
    texto = texto.casefold()
    texto = texto.replace("  ", " ")
    texto = texto.strip()
    return texto

#retornar um boolean
def bateu_meta(meta, vendas):
    if vendas >= meta:
        return True
    else:
        return False
    
#retornar uma lista, tupla ou dicionario
def filtrar_lista_texto(lista, pedaco_texto):
    lista_filtrada = []
    for item in lista:
        if pedaco_texto in item:
            lista_filtrada.append(item)
    return lista_filtrada
```

```python
#Exemplos:

# 1)-exemplo
nota1=8
nota2=9
nota3=7
soma= minha_soma(nota1,nota2,nota3)
print(soma)

#*************************************************************************

# 2)-Exemplo
texto_sem_padrao="o SoL naSCeU FORte naquela manhã, mas o veNTo ainda estava frio"

texto_filtrado=padronizar_texto(texto_sem_padrao)
print(texto_filtrado)

#*************************************************************************

# 3)-Exemplo
meta= 5000
vendas=7000

resultado=bateu_meta(meta,vendas)
print(resultado)

#**************************************************************************

#4)- Exemplo
listas_textos=["bruno@gmail.com","bruno@hotmail.com,","bruno@yahoo.com", "rocha@gmail.com","rocha@hotmail.com", "rocha@yahoo.com"]

lista= filtrar_lista_texto(listas_textos, "hotmail")
print(lista)
```
